```python
import sqlite3
import os

# Connect to the database
db_path = 'inventory.db'
if not os.path.exists(db_path):
    db_path = '/workspaces/spark-template/backend/inventory.db'

print(f"Connecting to {db_path}")
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Check existing columns
cursor.execute("PRAGMA table_info(trade_ins)")
columns = [info[1] for info in cursor.fetchall()]
print(f"Current columns: {columns}")

# Add 'color' column if it doesn't exist
if 'color' not in columns:
    print("Adding 'color' column...")
    cursor.execute("ALTER TABLE trade_ins ADD COLUMN color TEXT")
else:
    print("'color' column already exists.")

# Add 'capacidad' column if it doesn't exist
if 'capacidad' not in columns:
    print("Adding 'capacidad' column...")
    cursor.execute("ALTER TABLE trade_ins ADD COLUMN capacidad TEXT")
else:
    print("'capacidad' column already exists.")

conn.commit()

# Verify changes
cursor.execute("PRAGMA table_info(trade_ins)")
new_columns = [info[1] for info in cursor.fetchall()]
print(f"Updated columns: {new_columns}")

conn.close()
```
